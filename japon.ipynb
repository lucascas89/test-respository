```python
import requests
import json
import pandas as pd
```

```python
url = 'http://service.iris.edu/fdsnws/event/1/query?starttime=1995-01-01T00:00:00&endtime=2001-12-31T00:00:00&orderby=time&format=geocsv&lat=36.205&lon=138.253&minradius=0&maxradius=10.000&nodata=404'
r = requests.get(url)
data= r.text
file= open('JP9501.txt' , 'w')
file.write(data)
file.close
pd.read_csv('JP9501.txt' , delimiter='|' , skiprows=4 , low_memory = False)
```
